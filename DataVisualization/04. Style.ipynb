# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
x= [1,2,3]
y= [2,4,8]
```

```python
plt.plot(x,y)
plt.title('그래프')
plt.show()
```

```python
plt.plot(x,y, linewidth=5) #굵기
plt.plot(x,y, marker='o') # 점 표시
```

```python
plt.plot(x,y, marker ='o' , linestyle='None')
```

```python
plt.plot(x,y, marker ='v' , markersize='10')
```

```python
plt.plot(x,y, marker ='X' , markersize='10')
```

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y,linestyle="-.")
```

## 색상 수정 

```python
plt.plot(x,y,color='red')
```

##  포맷 

```python
plt.plot(x,y, 'ro--') #color, marker,linestyle
```

```python
plt.plot(x,y,'bv:')
```

## 축약어 

```python
plt.plot(x,y, marker ='o', mfc='red', ms=10, mec='blue', ls=':')
```

## 투명도 

```python
plt.plot(x,y, marker ='o', mfc='red', ms=10, alpha=0.2)
```
