# 6. 텍스트 

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
x=[1,2,3]
y=[2,4,8]
plt.plot(x,y)
```

```python
plt.plot(x,y,marker='o')
for idx, txt in enumerate(y): #설명은 밑에 참조 
    plt.text(x[idx],y[idx],txt)
```

```python
plt.plot(x,y,marker='o')
for idx, txt in enumerate(y): #enumerate: 인덱스, 값
    plt.text(x[idx],y[idx]+0.3,txt, ha='center') # x위치,y 위치, txt
```
