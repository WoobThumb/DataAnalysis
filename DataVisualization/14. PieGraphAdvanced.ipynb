# 14. 원 그래프(심화)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='AppleGothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
values=[30,25,20,13,10,2]
labels=['Python','Java','Javascript','C#','C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
values=[30,25,20,13,10,2]
labels=['Python','Java','Javascript','C#','C/C++', 'ETC']
colors=['b','g','r','c','m','y']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors)
plt.show()
```

```python
 wedgeprops={'width':0.6}

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, explode=[0.05]*6, wedgeprops=wedgeprops )
plt.show()
```

```python
 wedgeprops={'width':0.6, 'edgecolor':'w', 'linewidth':1.5}

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, explode=[0.05]*6, wedgeprops=wedgeprops )
plt.show() 
```

```python
def custom_autopct(pct):
    return('%.1f%%'%pct) if pct >=10 else ''
            #'{:.1f}%'.format(pct) if pct >=10 else ''}

plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, counterclock=False, explode=[0.05]*6, wedgeprops=wedgeprops)
plt.show() 
```

```python
plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, counterclock=False, explode=[0.05]*6, wedgeprops=wedgeprops, pctdistance= 0.7)
plt.show() 
```
