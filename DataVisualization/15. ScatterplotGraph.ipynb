# 15. 산점도 그래프 

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size']= 15
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
import pandas as pd
df= pd.read_csv('../Pandas/score.csv')
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000 #학생 수 8만큼 랜덤으로랜덤으로
sizes
```

```python
plt.scatter(df['영어'],df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수') 
```

```python
#학년별로 점 크기 다르게 하기

sizes= df['학년'] * 500

plt.scatter(df['영어'],df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수') 
```

```python
#학년별로 색상 다르게
plt.scatter(df['영어'], df['수학'], s= sizes, c=df['학년'], cmap='viridis', alpha=0.3)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['영어'], df['수학'], s= sizes, c=df['학년'], cmap='viridis', alpha=0.3) 
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['영어'], df['수학'], s= sizes, c=df['학년'], cmap='viridis', alpha=0.3) 
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1,2,3], label='학년', shrink=0.5)
```
